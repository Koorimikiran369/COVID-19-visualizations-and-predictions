```python
import pandas as pd
import numpy as np
import re
from bs4 import BeautifulSoup 
import requests
```

```python
website_url = requests.get('https://covid19ind.zaoapp.net/world/').text
```

```python
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
Title = soup.title.text
Title
```

```python
data_tag = soup.find_all('td')
data_tag
```

```python
Country_Name = []
Confirmed  = []
Recovered = []
Deaths = []

    
```

```python
Country_Name = []
for item in data_tag[0::4]:
    if len(item)==1:
        match = item.text

        Country_Name.append(match)
    else:
        Country_Name.append(np.nan)
```

```python
len(Country_Name)
```

```python
Total_confirmed = []
for item in data_tag[1::4]:
    if len(item)==1:
        match = item.text

        Total_confirmed.append(match)
    else:
        Total_confirmed.append(np.nan)
```

```python
len(Total_confirmed)
```

```python
Recovered = []
for item in data_tag[2::4]:
    if len(item)==1:
        match = item.text

        Recovered.append(match)
    else:
        Recovered.append(np.nan)
```

```python
len(Recovered)
```

```python
Deaths = []
for item in data_tag[3::4]:
    if len(item)==1:
        match = item.text

        Deaths.append(match)
    else:
        Deaths.append(np.nan)
```

```python
len(Deaths)
```

```python
Covid_data = pd.DataFrame()
Covid_data['Country_Name'] = Country_Name
Covid_data['Total_confirmed'] = Total_confirmed
Covid_data['Total_Recovered'] = Recovered
Covid_data['Total_Deaths'] = Deaths
```

```python
Covid_data
```

```python
Covid_data.Country_Name
```

```python
#re.sub('[^A-Za-z0-9]+', '', Covid_data.Country_Name.str)
Covid_data['Country_Name'] = Covid_data['Country_Name'].str.replace('\W', '')
```
